**IDENTAÇÃO DE CÓDIGO**

```python
def sacar(valor):
  saldo = 500

  if saldo >= valor:
    saldo -= valor
    print(f'Saque Realizado, saldo atual R${saldo}')

sacar(100)
```

**ESTRUTURAS CONDICIONAIS**

- if se
- elif se aninhado
- else

```python
# Verificador de CNH

idade = int(input('Digite sua idade: '))

if idade>=18:
  print('Voce ja pode tirar sua CNH')
else:
  print('Voce ainda nao pode tirar sua CNH')
```

```python
# Sistema de Saldo Bancário

saldo = 2000.0
saque = float(input('Digite o valor do saque: '))

if saque > saldo:
  print('Impossivel realizar o saque')
else:
  print('Saque realizado')
```

```python
opcao = int(input('Digite uma opcao: \n[1] saque \n[2] extrato\n'))

if opcao == 1:
  print('Saque')
elif opcao == 2:
  print('Extrato')
else:
  print('Opcao invalida')
```

```python
# if ternario

status = 'Sucesso' if saldo >= saque else 'Falha'
print(f'{status} ao realizar o saque')
```

**ESTRUTURAS DE REPETICÃO**

- for (usado para percorrer um objeto iterável)
- while
- range (start,stop,step) obs.: final -1 para percorrer todos os valores
- break (para encerrar o laço)

```python
texto = input('Digite um texto: ')
vogais = 'AEIOU' # STR é um objeto iterável

for letra in texto:
  if letra.upper() in vogais:
    print(letra, end=' ')
else:
  print()
  print('Fim do laço') # Adicionar uma quebra de linha
```

```python
opcao = -1

while opcao != 0:
  opcao = int(input('Digite uma opcao: \n[1] saque \n[2] extrato \n[0] sair\n'))

  if opcao == 1:
    print('Saque')
  elif opcao == 2:
    print('Extrato')
  elif opcao == 0:
    print('Saindo do programa, volte sempre')
  else:
    print('Opcao invalida')
```

```python
while True:
  numero = int(input('Digite um numero: '))
  if numero == 10:
    break
  print(numero)

print('Laço finalizado')
```

```python
for numero in range(100):
  if numero == 10:
    break

  print(numero)

print('Fim em 10')
```

```python
for numero in range(21):
  if numero % 2 == 0:
    print(numero,' par')
  else:
    print(numero,' impar')

```
