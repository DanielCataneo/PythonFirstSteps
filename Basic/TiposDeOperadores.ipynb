**OPERADORES ARITMÉTICOS**

Ordem de operação

- Parêntesis
- Expoêntes
- Multiplicao e divisao (esquerda --> direita)
- Soma e subtracao  (esquerda --> direita)

```python
num_1 = 10
num_2 = 20

# Adição
print(num_1 + num_2)

# Subtração
print(num_2 - num_1)

# Divisao
print(num_2 / num_1)

# Multiplicação
print(num_2 * num_1)

# Potenciação
print(num_2 ** num_1)

# Divisão Inteira
print(num_2 // num_1)

# Módulo
print(num_2 % num_1)
```

```python
print(num_1 + num_2 * 3)
print((num_1 + num_2) * 3)
```

**OPERADORES DE COMPARAÇÃO**
- '>' maior
- '<' menor
- '==' igual
- '!=' diferente
- '>=' maior igual
- '<=' menor igual

```python
saldo = 200
saque = 300

print(saldo > saque)
print(saldo < saque)
print(saldo == saque)
print(saldo != saque)
print(saldo >= saque)
print(saldo <= saque)
```

**OPERADORES DE ATRIBUIÇÃO**
- = atribui
- += adiciona
- -= diminui
- *= multiplica
- /= divide
- //= divide com numero inteiro
- **= exponenciacao

```python
saldo = 300

saldo += 200
print(saldo)

saldo -= 200
print(saldo)

saldo *= 2
print(saldo)

saldo /= 2
print(saldo)

saldo //= 2
print(saldo)

saldo **= 2
print(saldo)
```

**OPERADORES LÓGICOS**
- and &
- or |
- not !


```python
saldo = 1000
saque = 200
limite = 100
conta_especial = True

print(saldo >= saque and saldo >= limite)

print(saldo <= saque or saldo <= limite)

print(not 1000 > 1500)

# Utilizando Parenteses
(saldo >= saque and saque <= limite) or (conta_especial and saldo >= saque)

```

**OPERADORES DE ASSOCIAÇÃO**
- 'in' esta dentro
- 'not in' não esta dentro

```python
curso = 'Curso de Python'
frutas = ['laranja','uva','limao']
saques = [1500,100]

print('Python' in curso)

print('maça' not in frutas)

print(101 in saques)
```
